# Opis

- problem skalowania wielkości sygnału danych - przeskalowane przez stałą odpowiedniego rzędu (np. $10^{-7}$). 


- wykluczenie problemów naturalnie występującego "slow drift" oraz ograniczenie się do standardowych zakresów częstoliwości fal występujących w mózgu - FIR filtr w zakresach $[0,1]$, $[90, \infty]$


- Interferencja fal przy 50 hz - nałożony Notch Filter

- Wykorzystanie do klasyfikacji - Random Forest Classifier
