## EEG Data Info

```python
filename = r'datasets\2D663E30\EEG_ExperimentBlockDECEITFUL_RESPONSE_TO_FAKE_IDENTITY_raw.fif'
raw = mne.io.read_raw_fif(fname=filename)

print(raw.info)

print(raw.ch_names)
```

```python
raw.crop(tmin=4.0, tmax=8.0)

data, times = raw.get_data(return_times=True)
 
data = data * 1e6
```

```python
import matplotlib.pyplot as plt

plt.plot(times, data[0])
plt.show()
```
